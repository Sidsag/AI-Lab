```python
#Question 1
#Load the csv file of the dataset into a dataframe in Python program.
import pandas as pd
#load dataframe from csv
#df = pd.read_csv(data_url)
data_url='https://raw.githubusercontent.com/fivethirtyeight/data/master/us-weather-history/KCLT.csv'
df = pd.read_csv(data_url)
#print dataframe
print(df)
```

```python
#Question 2
#Find the mean, median, mode, min and max for all numeric attributes.
numeric_ColumnDf=df[["actual_mean_temp","actual_min_temp","actual_max_temp","average_min_temp","average_max_temp","record_min_temp","record_max_temp","actual_precipitation","average_precipitation","record_precipitation"]].copy()
print(numeric_ColumnDf)
print("\n----------- Calculate Mean -----------\n")
print(numeric_ColumnDf.mean())
 
print("\n----------- Calculate Median -----------\n")
print(numeric_ColumnDf.median())
 
print("\n----------- Calculate Mode -----------\n")
print(numeric_ColumnDf.mode())
```

```python
#Question 3
#Print the top 20% of rows showing only the first four columns.
n=len(df)*0.2
df.iloc[:, : 4].head(int(n))
```

```python
#Question 4
#Create a new column (call it ‘newColumn’), it should have the same values as the column‘actual_mean_temp’. Print head of dataframe.
df['newColumn'] = df['actual_mean_temp']
df.head()
```

```python
#question 5
#Remove the new column that you have created above. Print head of dataframe.
del df['newColumn']
df.head()
```

```python
#Question 6
#Print the first 10 rows....save row containing data of ‘2014-7-3’  in a variable of type series (data structure)
print(df.iloc[0:10])
ser = df.loc[df['date'] == '2014-7-3']
print("The row with date 2014-7-3 is: ")
print(ser)
df = df[df.date != '2014-7-3']
print(df.iloc[0:10])
```

```python
#Question 7
## Add the row that you deleted before. Print the first 10 rows again.
df = pd.concat([df, ser], ignore_index = True)
print(df.iloc[0:10])
```

```python
#Question 8
#Update the actual_min_temp in the data row for date ‘2014-7-3’ to any value. Print the updated row.
#nm=df['date' == '2014-7-3'].index
for index, row in df.iterrows():
    if df.loc[index,'date'] == '2014-7-3':
       df.at[index,'actual_min_temp']=90
show=df[df.date=='2014-7-3']
print(show)
```

```python
#Question 9
# Print these 10 rows before and after the operation.
df.head(10)
```

```python
#Question 9-----continue
##Add 5 to the ‘actual_mean_temp’ wherever the ‘actual_min_temp’ is odd. Do this only on the top 10 rows. Print these 10 rows before and after the operation.
for index, row in df.head(10).iterrows():
    if df.loc[index,'actual_min_temp']%2!=0:
       df.at[index,'actual_mean_temp']=df.at[index,'actual_mean_temp']+5
df.head(10)
```

```python
#Question 10
#Print only those rows where the absolute difference between the ‘record_min_temp_year’ and ‘record_max_temp_year’ is less than 30.
for index, row in df.iterrows():  
    if abs(df.loc[index,'record_max_temp_year'] - df.loc[index,'record_min_temp_year']) < 30:
        print(row)
```
